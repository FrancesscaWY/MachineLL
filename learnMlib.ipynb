```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

### 1.1 折线图

```python
# 创建画布
plt.figure(figsize=(20,8),dpi=100)
# 绘制图像
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y)
# 保存图像
# plt.savefig("/home/francessca/Pictures/pic/plot.png")
# 显示图像 
plt.show() #  show函数显示后释放资源
```

- 案例学习 -- 温度变化折线图

```python
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(15,25) for i in x]

plt.figure(figsize=(20,8),dpi=100)
# plt.plot(x,y_shanghai,label="ShangHai")
# plt.plot(x,y_beijing,label="BeiJing")
fig,axes = plt.subplots(nrows = 1 ,ncols = 2, figsize=(20,8),dpi = 100)

axes[0] .plot(x,y_shanghai,label = "ShangHai")
axes[1] .plot(x,y_beijing,label = "BeiJing")
y_ticks = range(30)
x_ticks_labels = ["11 : {}".format(i) for i in x]

axes[0].set_yticks(y_ticks[12::1])
axes[1].set_yticks(y_ticks[12::1])
axes[0].set_xticks(x[::5],x_ticks_labels[::5])
axes[1].set_xticks(x[::5],x_ticks_labels[::5])
axes[0].grid(True,linestyle='--',alpha=0.5)
axes[1].grid(True,alpha =0.5)
axes[0].set_xlabel("Time")
axes[1].set_xlabel("Time")
axes[0].set_ylabel("Temperature")
axes[1].set_ylabel("Temperature")
axes[0].set_title("Change of Temperature an Hour in ShangHai",fontsize = 20)
axes[1].set_title("Change of Temperature an Hour in Beijing",fontsize = 20)
axes[0].legend(loc = "best")
axes[1].legend(loc = "best")
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.sin(x)

plt.figure(figsize=(20,8),dpi =100)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

### 1.2离散图

```python
datanum = range(100)
area = [random.uniform(200,400) for i in datanum]
price = [random.uniform(1000,1500) for i in datanum]
plt.figure(figsize=(20,8),dpi =100)
plt.scatter(area,price,color = "g")
plt.xlabel("Area",fontsize = 20)
plt.ylabel("Price",fontsize = 20)
plt.title("The Ralation of House Area and Price")
plt.show()
```
